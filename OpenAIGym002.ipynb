```python
'''
OpenAIGym002: Open AI Gymの基本的な使い方（山登り課題）

[Note]
* 事前に OpenAIGym001 が正常に動作することを確認しておくこと
'''

###################################################################
#
# 下準備
#
###################################################################

# Open AI Gymモジュールを読み込む
import gym

# "10x10グリッド迷路"環境を読み込む
env = gym.make('MountainCar-v0')  # <-- RLex001からこの1文を変更

# 環境の初期化
env.reset()

###################################################################
#
# 1エピソード（1試行）内の基本的な流れ
#
###################################################################

# 1試行内の最大タイプステップを定義する
MaxT = 1000

# 各時間ステップごとにfor文内の処理を繰り返す
for t in range(1000):
    
    # 現在の課題の状態を描画する
    env.render()
    
    # ランダムアクションを生成する
    action = env.action_space.sample()
    
    # actionで指定された行動を環境に渡して、1時間ステップ進める
    done = env.step(action)
    

###################################################################
#
# 後処理
#
###################################################################    

# 環境を閉じる
env.close()
```
