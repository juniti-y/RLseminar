```python
'''
OpenAIGym001: Open AI Gymの基本的な使い方（CartPole課題）

[Note]
* pyglet, pygameライブラリ（描画用ライブラリ）がインストールされていないとエラーになる
* pygletのインストール方法: プロンプトより以下を入力
  % pip install pyglet 
* pygameのインストール方法: プロンプトより以下を入力
  % pip install pygame
'''

###################################################################
#
# 下準備
#
###################################################################

# Open AI Gymモジュールを読み込む
import gym

# "CartPole"環境を読み込む（引数を変えることで様々な課題を環境にセットできる）
env = gym.make('CartPole-v1')

# 環境の初期化
env.reset()

###################################################################
#
# 1エピソード（1試行）内の基本的な流れ
#
###################################################################

# 1試行内の最大タイプステップを定義する
MaxT = 1000

# 各時間ステップごとにfor文内の処理を繰り返す
for t in range(1000):
    
    # 現在の課題の状態を描画する
    env.render()
    
    # ランダムアクションを生成する
    action = env.action_space.sample()
    
    # actionで指定された行動を環境に渡して、1時間ステップ進める
    done = env.step(action)
    

###################################################################
#
# 後処理
#
###################################################################    

# 環境を閉じる
env.close()
```
