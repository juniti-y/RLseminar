```python
'''
OpenAIGym003: Open AI Gymの基本的な使い方（離散迷路課題）

[Note]
* Github https://github.com/MattChanTK/gym-maze からパッケージをダウンロード
* gym-mazeのインストール方法:  プロンプトより以下を入力
  % cd gym-maze （ダウンロード先のフォルダ: そのフォルダにsetup.pyが存在するか要確認）
  % python setup.py install
* インストール後はカーネルの再起動が必要
'''

###################################################################
#
# 下準備
#
###################################################################

# Open AI Gymモジュールを読み込む
import gym

# Open AI Gym用グリッド迷路課題ライブラリを読み込む
import gym_maze  # <-- RLex002にこの1文を追加

# "10x10グリッド迷路"環境を読み込む
env = gym.make('maze-sample-10x10-v0')  # <-- RLex002からこの1文を変更

# 環境の初期化
env.reset()

###################################################################
#
# 1エピソード（1試行）内の基本的な流れ
#
###################################################################

# 1試行内の最大タイプステップを定義する
MaxT = 1000

# 各時間ステップごとにfor文内の処理を繰り返す
for t in range(1000):
    
    # 現在の課題の状態を描画する
    env.render()
    
    # ランダムアクションを生成する
    action = env.action_space.sample()
    
    # actionで指定された行動を環境に渡して、1時間ステップ進める
    done = env.step(action)
    

###################################################################
#
# 後処理
#
###################################################################    

# 環境を閉じる
env.close()
```
